```python
# -*- coding: utf-8 -*-
"""
Created on Tue Jan 15 17:25:11 2019

@author: kinde
"""

import pandas as pd
import numpy as np

from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing



top5000OverAllArtists = pd.read_csv("top5000_artist_overall_tfidf.csv", sep="\t")
```

```python
top5000Gender = pd.read_csv("top5000_artist_gender_tfidf.csv", sep="\t")
```

```python
DropArtistOverAll = top5000OverAllArtists.drop('user_id', axis=1)
DropArtistGender = top5000Gender.drop('user_id', axis=1)

print(DropArtistOverAll.head(), DropArtistOverAll.shape)
print(DropArtistGender.head(), DropArtistGender.shape)
```

```python
#merging them:
result = pd.concat([DropArtistOverAll, DropArtistGender], axis=1, sort=False)
print(result.shape, result.head())
```

```python
#PCA of "result" DF:
from sklearn.decomposition import PCA
pca = PCA(n_components=450)
pca_result = pca.fit_transform(result)

print(pca_result)
```

```python
new_df_without_users = pd.DataFrame(data = pca_result)
print (new_df_without_users.head())
```

```python
#adding the users:
new_df_without_users['user_id'] = top5000Gender['user_id'].values
print (new_df_without_users.head())
```

```python
# now i need to add the user info
UserInfo = pd.read_csv("LFM-1b_subset_MTAP2018.csv", sep='\t')
UserInfo.columns =['user_id','age','gender','country']
merged_df = pd.merge(UserInfo, new_df_without_users, on='user_id')

print(new_df_without_users.shape)
print(merged_df.head(), new_df_without_users.shape)
```

```python
malesOnly = merged_df.loc[merged_df.gender == "m"][:2545]


females = merged_df.loc[merged_df.gender == "f"]

merged_df = pd.concat([malesOnly, females])
print(merged_df.head(), merged_df.shape)
```

```python
#importing user stats and merging them:
user_stats = pd.read_csv("LFM-1b_users_additional.txt", sep='\t')
merged_df2 = pd.merge(merged_df, user_stats, on='user_id')

print(merged_df2.head(),merged_df2.shape)

```
